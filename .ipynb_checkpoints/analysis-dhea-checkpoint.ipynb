```python
import pandas as pd
import numpy as np
from IPython.display import display
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('cleandate.csv')
data.head(5)
```

```python
data.dtypes
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.dtypes
```

```python
a = sns.displot(data, x="subject", bins=4)
a.set_xticklabels(rotation=45)
plt.show()
```

```python
totals = []
dates = data.Date.unique()
dates.sort()
for i in dates:
    tempDF = data[data['Date']==i]
    totals.append([i,len(tempDF)])
plt.figure(figsize=(12,6))
sns.regplot(x=[i for i in range(len(totals))],
            y=[i[1] for i in totals],
            scatter_kws={'color': '#A435F0'}, line_kws={'color': '#199FA3'})
plt.title(label='Course uploads over time (Business + Web + Design + Music)')
plt.xlabel(xlabel='Time (from 2011 to 2017)')
plt.ylabel(ylabel='Number of uploads')
plt.xticks([])
plt.show()
```
